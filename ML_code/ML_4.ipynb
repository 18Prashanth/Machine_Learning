```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('/content/insurance_data.csv')
df = pd.DataFrame(data)
```

```python
df
```

```python
df.head()
```

```python
plt.scatter(df.age, df.bought_insurance, color='red', marker='+')
```

```python
model = linear_model.LinearRegression()
model.fit(df[['age']], df.bought_insurance)
plt.scatter(df.age, df.bought_insurance, color='red', marker='+')
plt.plot(df.age, model.predict(df[['age']]), color='red')
```

```python
#For training and test we devide the data set
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size=0.9, random_state=10)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test, y_test)
```
